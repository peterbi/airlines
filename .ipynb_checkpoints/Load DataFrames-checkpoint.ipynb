```python
def fillna_col(series):
    if series.dtype is pd.np.dtype(object):
        return series.fillna('')
    else:
        return series
```

```python
def nullify(value):
    if value =="\\N":
        return float('nan')
    else:
        return value
```

```python
performance_dfn=performance_df.applymap(nullify)
```

```python
airlines_dfn=airlines_df.applymap(nullify)
```

```python
airports_dfn=airports_df.applymap(nullify)
```
