```python
airline_cities_df = pd.read_sql(' with ct (route_airline_id, num) as (select route_airline_id, count(distinct(dst_apt_cd)) as \
a from routes group by route_airline_id order by a) \
select t2.airline_name, t1.num \
from airlines t2 join ct t1 on t2.airline_id=t1.route_airline_id where num>200 order by num desc', db)
```

```python
airline_cities_df
```

```python
airline_cities_df.plot(kind= 'bar', x='airline_name', y='num', title= 'Number of Destinations by Airline',
                       legend= True, grid=True, fontsize=12) 
```

```python
plt.show()
plt.savefig('airlineHistogram.svg') 
```

```python
route_c=routes_df.groupby('route_airline_id')['dst_apt'].nunique()
df=pd.DataFrame(route_c)
code_dst=df[df['dst_apt']>200]
#.sort_values('dst_apt',ascending=False)
num_fj=pd.merge(airlines_df, code_dst, left_on = "airline_id", right_index = True)
p_airlines_cities_df=num_fj[['airline_name','dst_apt']]
p_airlines_cities_df.columns=['airline_name','num']
p_airlines_cities_df=p_airlines_cities_df.sort_values('num',ascending=False)
p_airlines_cities_df
```

```python
nonstop_df2 = routes_df[['route_airline_id','src_apt_cd','dst_apt_cd']]
nonstop_df2.columns=['airline_id','source_id','target_id']
nonstop_df2.to_sql('nonstop',con=db,if_exists='replace',index=False)
```

```python
nonstop_df2
```

```python
onestop_df = pd.read_sql('select t1.airline_id, t1.source_id, t2.target_id from nonstop t1 join nonstop t2 on t1.airline_id=t2.airline_id and t1.target_id=t2.source_id and t1.source_id<>t2.target_id  ', db)
```

```python
onestop_df
```

```python
onestop_df.to_sql('onestop',con=db,if_exists='replace',index=False)
twostop_df = pd.read_sql('select t1.airline_id, t1.source_id, t2.target_id from nonstop t1 join onestop t2 \
on t1.airline_id=t2.airline_id and t1.target_id=t2.source_id and t1.source_id<>t2.target_id, order by airline_id ', db)
```

```python
twostop_df
```

```python
nonstop_df2=nonstop_df2.assign(num_stops = 0)
onestop_df=onestop_df.assign(num_stops = 1)
twostop_df=twostop_df.assign(num_stops = 2)
```

```python
nonstop_df2
```

```python
two_or_fewer_stops_df=pd.concat([nonstop_df2,onestop_df,twostop_df],ignore_index=True)
```
